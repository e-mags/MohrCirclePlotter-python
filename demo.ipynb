```python
import numpy as np

from mohrcircleplotter import MohrPlot
import matplotlib.pyplot as plt


mp = MohrPlot(show_tension=True, halfplot=True)

mp.principal_stress_circle(p1=380, p2=0, label='Circle 1')
mp.principal_stress_circle(p1=401.4, p2=3, label='Circle 2')
mp.principal_stress_circle(p1=427.4, p2=7, label='Circle 3')
mp.principal_stress_circle(p1=436.2, p2=10, label='Circle 4')

mp.estimated_mc_envelope(parameters_label=True)
mp.mc_envelope(phi=np.radians(30), c=90, parameters_label=True)

mp.oriented_circle(sigma_x=100, sigma_y=380, tau_xy=50, label='Oriented Circle 1')
plt.legend(bbox_to_anchor=(1.05, 0.5), loc='upper left')
```

```python

from mohrcircleplotter import MohrPlot

# Create a plotter instance which contains a matplotlib figure and axis
plot = MohrPlot(halfplot=True,show_tension=True, units='MPa')

# Define some circles (sigma_x, sigma_y, tau_xy)
circles = [
    (80, 30, 30),
    (60, 20, 20),
    (40, 10, 10)
]

# Plot the circles with red dashed lines showing the orientation of $$\sigma_x$$
for sigma_x, sigma_y, tau_xy in circles:
    plot.oriented_circle(sigma_x, sigma_y, tau_xy)
```

```python
from mohrcircleplotter import MohrPlot

plot = MohrPlot(halfplot=True,show_tension=True, units='MPa')

# Define principal stresses (sigma_1, sigma_2)
principal_stresses = [
    (90, 50),
    (60, 30),
    (40, 10)
]

# Plot circles from principal stresses
for sigma_1, sigma_2 in principal_stresses:
    plot.principal_stress_circle(sigma_1, sigma_2)
```

```python
from mohrcircleplotter import MohrPlot

plot = MohrPlot(halfplot=True,show_tension=True, units='MPa')

# Define some circles (sigma_x, sigma_y, tau_xy)
circles = [
    (80, 30, 30),
    (60, 20, 20),
    (40, 10, 10)
]

for sigma_x, sigma_y, tau_xy in circles:
    plot.oriented_circle(sigma_x, sigma_y, tau_xy)

# Estimate and plot the Mohr-Coulomb failure envelope
plot.estimated_mc_envelope()
```

```python

import matplotlib.pyplot as plt
from mohrcircleplotter import oriented_circle

# Create a new figure and axis
fig, ax = plt.subplots()

# Define some circles (sigma_x, sigma_y, tau_xy)
circles = [
    (80, 30, 30),
    (60, 20, 20),
    (40, 10, 10)
]

# Plot the circles on the current axis
for sigma_x, sigma_y, tau_xy in circles:
    oriented_circle(sigma_x, sigma_y, tau_xy, ax=ax)

plt.gca().set_aspect('equal', adjustable='box')
plt.grid()
```
